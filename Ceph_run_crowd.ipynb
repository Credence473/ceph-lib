### Load files

```python
file = "data/jw01685-c1007_t003_nircam_clear-f150w_i2d.fits"
coord_tab = "data/Yuan_4258_2022_culled_good.csv"

photometry = ceph_phot(file, coord_tab, r=0.15, n_stars=10)
crowd = True
photometry.get_psf()
```

```python
# photometry.plot_psf()
```

```python
photometry.do_photometry(ap_size=25, do_crowd=crowd, verbose=True)
```

```python
op = photometry.results()
plots = ceph_plot(file, op, plot_crowd=crowd)
```

```python
x = plots.table["ra_pix"]
y = plots.table["dec_pix"]
z = plots.table["b"]

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20, 10))
ax.imshow(photometry.data, cmap='viridis', origin='lower',vmin=0.01, vmax=1)
im = ax.scatter(x, y, s=5, c=z, cmap='inferno_r')
plt.colorbar(im, ax=ax)
plt.show()
```
